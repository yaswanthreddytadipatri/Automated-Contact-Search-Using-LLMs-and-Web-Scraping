```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin, urlparse
import time

# Initialize a set to keep track of visited URLs
visited_urls = set()
data = {}

def scrape_website(url, max_depth=2, current_depth=0):
    # If the maximum depth is reached, stop recursion
    if current_depth > max_depth:
        return

    # Check if the URL has already been visited
    if url in visited_urls:
        return

    # Mark the URL as visited
    visited_urls.add(url)

    try:
        # Send a GET request to the URL
        response = requests.get(url, timeout=10)

        # Check if the request was successful
        if response.status_code != 200:
            print(f"Failed to retrieve {url}. Status code: {response.status_code}")
            return

        # Parse the content of the response with BeautifulSoup
        soup = BeautifulSoup(response.content, 'html.parser')

        # Print the title of the page
        print(f"Title: {soup.title.string if soup.title else 'No title'} - URL: {url}")
        tags.append(f"Title: {soup.title.string if soup.title else 'No title'} - URL: {url}")
        data[soup.title.string if soup.title else None] = url
        full_data.append(soup)
        # Find all links on the page
        links = soup.find_all('a','')

        # Recursively scrape each link found
        for link in links:
            href = link.get('href')
            if href:
                # Join the URL if it's relative
                full_url = urljoin(url, href)
                # Only scrape links within the same domain
                if urlparse(url).netloc == urlparse(full_url).netloc:
                    scrape_website(full_url, max_depth, current_depth + 1)
    except Exception as e:
        print(f"An error occurred while scraping {url}: {e}")

# Start scraping from the initial URL
start_url = input("Enter the website: ")
tags = []
full_data = []
scrape_website(start_url, max_depth=1)
# print(tags)

# print(data)
# print(full_data)
```
